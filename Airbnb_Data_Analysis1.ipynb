# 🏡 Airbnb Data Analysis Project

This project performs an exploratory data analysis (EDA) on Airbnb listings dataset to understand various trends affecting pricing, availability, and locations.

```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set style
sns.set(style="whitegrid")
%matplotlib inline
```

```
# Load the Airbnb dataset
df = pd.read_csv('/mnt/data/airbnb_data/Airbnb_Data.csv')
df.head()
```

```
# Dataset Info
df.info()
```

```
# Summary statistics
df.describe(include='all')
```

```
# Check for missing values
df.isnull().sum().sort_values(ascending=False)
```

```
# Drop rows/columns with excessive missing data (example)
df_clean = df.dropna(thresh=0.7*len(df), axis=1)
df_clean = df_clean.dropna()
df_clean.info()
```

```
# Price distribution
plt.figure(figsize=(10,5))
sns.histplot(df_clean['price'], bins=50, kde=True)
plt.title('Price Distribution')
plt.xlabel('Price')
plt.ylabel('Number of Listings')
plt.xlim(0, df_clean['price'].quantile(0.95))
plt.show()
```

```
# Listings by neighborhood
plt.figure(figsize=(12,6))
top_neighborhoods = df_clean['neighbourhood'].value_counts().head(10)
sns.barplot(x=top_neighborhoods.values, y=top_neighborhoods.index)
plt.title('Top 10 Neighborhoods by Number of Listings')
plt.xlabel('Number of Listings')
plt.ylabel('Neighborhood')
plt.show()
```

```
# Availability
plt.figure(figsize=(10,5))
sns.histplot(df_clean['availability_365'], bins=50, kde=False)
plt.title('Availability Distribution')
plt.xlabel('Days Available per Year')
plt.ylabel('Number of Listings')
plt.show()
```

```
# Correlation matrix
plt.figure(figsize=(10,6))
sns.heatmap(df_clean.corr(numeric_only=True), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix')
plt.show()
```

## 📌 Conclusion

- The price distribution is highly skewed. Most listings fall within a lower price range.
- Certain neighborhoods dominate the listing count.
- Availability varies widely, with many listings available year-round.
- Correlations show how various numerical features relate to price and availability.
